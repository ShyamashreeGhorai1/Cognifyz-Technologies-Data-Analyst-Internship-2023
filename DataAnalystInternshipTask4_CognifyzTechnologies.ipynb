```python
from scipy.stats import f_oneway
```

```python
# H0 : Weather and Popularity are not correlated with each other.
CategoryGroupLists= df.groupby('Weather')['Popularity'].apply(list)
AnovaResults = f_oneway(*CategoryGroupLists)
print('P-Value for Anova is: ', AnovaResults[1])
```

```python
pvalue = AnovaResults[1]
alpha = 0.05
if pvalue < alpha:
  print('we reject the null hypothesis')
else:
  print('we accept the null hypothesis')
```

```python
# Hence, we reject H0. Which means the variables are correlated with each other.
```

```python
# H0 : Artist  and Popularity are not correlated with each other
CategoryGroupLists= df.groupby('Artist')['Popularity'].apply(list)
AnovaResults = f_oneway(*CategoryGroupLists)
print('P-Value for Anova is: ', AnovaResults[1])
```

```python
pvalue = AnovaResults[1]
alpha = 0.05
if pvalue < alpha:
  print('we reject the null hypothesis')
else:
  print('we accept the null hypothesis')
```

```python
# Hence, we reject H0. Which means the variables are correlated with each other.
```

```python
# H0 : Album and popularity are not correlated with each other
CategoryGroupLists= df.groupby('Album')['Popularity'].apply(list)
AnovaResults = f_oneway(*CategoryGroupLists)
print('P-Value for Anova is: ', AnovaResults[1])
```

```python
pvalue = AnovaResults[1]
alpha = 0.05
if pvalue < alpha:
  print('we reject the null hypothesis')
else:
  print('we accept the null hypothesis')
```

```python
# Hence, we reject H0. Which means the variables are correlated with each other.
```

```python
# H0 : Album and popularity are not correlated with each other
CategoryGroupLists= df.groupby('Track Name')['Popularity'].apply(list)
AnovaResults = f_oneway(*CategoryGroupLists)
print('P-Value for Anova is: ', AnovaResults[1])
```

```python
# Hence there is no correlation between Track Name and Popularity.
```

Observation:
1. There is a correlation between:-

 i)Weather and Popularity,

 ii)Artist and Popularity and
